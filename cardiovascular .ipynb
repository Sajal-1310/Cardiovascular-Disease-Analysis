```python
# 13. Remove Outliers using IQR Method
# (Removing extreme values to make data cleaner)
def remove_outliers(df, col):
    Q1 = df[col].quantile(0.25) 
    Q3 = df[col].quantile(0.75)  
    IQR = Q3 - Q1  # 
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    return df[(df[col] >= lower) & (df[col] <= upper)] 
```
